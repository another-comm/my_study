str.maketrans(x, y)（两个参数）：

x 和 y 必须是等长的字符串，表示 x 中的每个字符会被替换成 y 中对应位置的字符。

```python
str.maketrans(x, y, z)（三个参数）：

x 和 y 仍然是等长的字符串，表示字符替换。

z 是一个字符串，表示要删除的字符（不会替换，直接删除）。
```

```python
trans = str.maketrans("abc", "123", "xyz")
print("a b c x y z".translate(trans))
```

# 问题解决：凯撒加密，每个字母替换为后面第k个。


```python
import string
def kaisa(s, k):
    lower = string.ascii_lowercase          #小写字母
    upper = string.ascii_uppercase          #大写字母
    before = string.ascii_letters
    after = lower[k:] + lower[:k] + upper[k:] + upper[:k]
    table = ''.maketrans(before, after)     #创建映射表
    return s.translate(table)
```

```python
s = "Python is a greate programming language. I like it!"
kaisa(s, 3)
```

## 问题解决: 请编写一个名为 StringProcessor 的类，要求具备以下功能：

> ### 构造函数: 接收一个字符串参数 text，并将其保存为实例属性。

> ### 实例方法：反转字符串: 定义方法 reverse()，返回原字符串的完全反转形式（如 "hello" 反转为 "olleh"）。

> ### 实例方法：统计子符串出现次数（不区分大小写）
定义方法 count_substring(sub)，返回子串 sub 在原字符串中出现的次数（不区分大小写）。

> ### 实例方法：去除标点符号
定义方法 remove_punctuation()，返回移除所有标点符号后的新字符串。标点符号包括：!"#$%&'()*+,-./:;<=>?@[\]^_{|}~。

> ### 类方法：从文件创建实例
定义类方法 from_file(cls, filename)，读取指定文件的内容，并用其内容创建一个 StringProcessor 实例。

> ### 类方法：对文件内容进行凯撒加密
定义类方法 encrypt(cls, text, shift)，对给定的 text 进行 凯撒加密（Caesar Cipher），偏移量为shift，并返回加密后的字符串。

```python
with open ('sample.txt','w',encoding='utf8') as fp:
    fp.write('Hello, World! This is a test.')
```

```python
import string

class StringProcessor:
    def __init__(self, text):
        self.text = text
    
    def reverse(self):
        return self.text[::-1]
    
    def count_substring(self, sub):
        main_str = self.text.lower()
        target = sub.lower()
        return main_str.count(target)
    
    def remove_punctuation(self):
        translator = str.maketrans('','', string.punctuation)
        return self.text.translate(translator)
    
    @classmethod
    def encrypt(cls, text, shift): 
        shift = shift
        lower = string.ascii_lowercase          #小写字母
        upper = string.ascii_uppercase          #大写字母
        before = string.ascii_letters
        after = lower[shift:] + lower[:shift] + upper[shift:] + upper[:shift]
        trans_table = str.maketrans(before, after)
        return text.translate(trans_table)
    
    @classmethod
    def from_file(cls, filename):
        with open(filename, 'r') as f:
            content = f.read().strip()
        return cls(content)
```

```python
sp = StringProcessor.from_file("sample.txt")
print(sp.text)                      # 输出: "Hello, World! This is a test."
print(sp.reverse())                 # 输出: ".tset a si sihT !dlroW ,olleH"
print(sp.count_substring("is"))     # 输出: 2
print(StringProcessor.encrypt("Khoor, Zruog!", shift=-3)) 
```
