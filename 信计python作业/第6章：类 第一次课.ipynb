## 2.  “面向过程”的实现方式

### 一步步的对需要的数据以及方法进行操作，根据业务逻辑从上到下写代码

```python
# 遍历打印所有姓名-年龄
def print_info(names_temp, ages_temp):
    print("姓名:", names_temp, " 年龄:", ages_temp)


# 定义一个变量存储姓名
name = "May"
# 定义一个变量存储年龄
age = 20

# 遍历姓名-年龄
print_info(name, age)
```

## 3. “面向对象”方式实现

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 遍历打印所有姓名-年龄
    def print_info(self):
        print('姓名:', self.name, '年龄:', self.age)


# 创建一个对象
p = Person('May', 20)
p.print_info()
```

# 利用类数据成员的共享性，可以实时获得该类的对象数量，并且可以控制该类可以创建的对象最大数量。例如：


```python
 class Demo(object):
    total = 0
    def __new__(cls, *args, **kwargs):           #该方法在__init__()之前被调用
        if cls.total >= 3:                       #最多允许创建3个对象
            raise Exception('最多只能创建3个对象')
        else:
            return object.__new__(cls)
    def __init__(self):
        Demo.total = Demo.total + 1
```

```python
class MyClass:
    def __init__(self):
        self.__private_var = 10  # 私有变量

    def __private_method(self):
        print("这是私有方法，访问私有变量:", self.__private_var)

    def public_method(self):
        print("公有方法调用私有方法:")
        self.__private_method()  # 在类内部通过 self 调用

obj = MyClass()
obj.public_method()  # 间接调用私有方法

# 外部强制调用私有方法（不推荐！）
obj._MyClass__private_method()
```

### 静态方法的理解（区别于实例方法和类方法）

```python
class Root:
    __total = 0
    height = 2
    def __init__(self, v):    #构造方法
        self.__value = v
        Root.__total += 1

    def show(self):           #普通实例方法
        print('self.__value:', self.__value)
        print('Root.__total:', Root.__total)
        staticShowTotal()
        
    @classmethod              #修饰器，声明类方法
    def classShowTotal(cls,height):  #类方法 def classShowTotal(cls,height):
        #cls.height = 5
        cls.height = height
        print(cls.__total)
            #修饰器，声明静态方法

def staticShowTotal():    #静态方法
    print(4)
```

```python
import abc
class Foo(metaclass=abc.ABCMeta):  #抽象类 ABCMeta是一个元类（metaclass），用于定义抽象基类的元信息
    def f1(self):                  #普通实例方法
        print(123)

    def f2(self):                  #普通实例方法
        print(456)

    @abc.abstractmethod            #抽象方法
    def f3(self):
        raise Exception('You musr reimplement this method.')
        
    @abc.abstractmethod            #抽象方法
    def f4(self):
        print(789)   

class Bar(Foo):
    def f3(self):                  #必须重新实现基类中的抽象方法
        print(33333)
    def f4(self):                  
        print('Hello')
```
