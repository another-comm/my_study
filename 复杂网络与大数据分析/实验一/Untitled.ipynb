```python
import numpy as np
import warnings

import pandas as pd
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore", category=UserWarning, message=".*normed.*")
import seaborn as sns
from scipy import stats
from scipy.stats import norm
%matplotlib inline
```

```python
df_train=pd.read_csv('./dataset/train.csv')
df_train.columns
```

```python
df_train['SalePrice'].describe()
```

```python
sns.histplot(df_train['SalePrice'], kde=True);
```

```python
print('偏度: {:.2f}'.format(df_train['SalePrice'].skew()))
print('峰度: {:.2f}'.format(df_train['SalePrice'].kurt()))
```

```python
# P1.6 GrLivArea 与 SalePrice关系
var = 'GrLivArea'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

```python
# P1.8 OverallQual 与 SalePrice关系
var = 'OverallQual'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 5))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
# P1.10 相关矩阵
# 加上 numeric_only=True 防止处理包含字符串的列时报错
corrmat = df_train.corr(numeric_only=True) 
# 将过时的 np.bool 替换为原生的 bool
mask = np.zeros_like(corrmat, dtype=bool) 
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corrmat, mask=mask, cmap="coolwarm", center=0,
            square=True, linewidths=.2, cbar_kws={"shrink":.8});

# P1.11 最相关矩阵
k = 10 # 只显示十组变量
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
# 同样将 np.bool 替换为 bool
mask = np.zeros_like(cm, dtype=bool) 
f, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm, mask=mask, annot=True, fmt='.2f', cmap="coolwarm", 
            vmax=1, vmin=-1, square=True, linewidths=.2, annot_kws={'size': 12}, 
            yticklabels=cols.values, xticklabels=cols.values);
```
