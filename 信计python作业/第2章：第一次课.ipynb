## +运算符除了用于算术加法以外，还可以用于列表、元组、字符串的连接，但不支持不同类型的对象之间相加或连接

```python
[1,2,3]+[5,7,9]
```

```python
(1,2,3)+(4,)
```

```python
True+1
```

```python
{1,2,3}-{3,4,5}
```

## *运算符除了表示算术乘法，还可用于列表、元组、字符串这几个序列类型与整数的乘法，表示序列元素的重复，生成新的序列对象。字典和集合不支持与整数的相乘，因为其中的元素是不允许重复的。

```python
[1,2,3]*3
```
