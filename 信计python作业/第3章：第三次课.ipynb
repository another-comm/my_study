## 3.3.3  元素添加、修改与删除

> ## 当以指定“键”为下标为字典元素赋值时，有两种含义：

> ## 1）若该“键”存在，则表示修改该“键”对应的值；

> ## 2）若不存在，则表示添加一个新的“键:值”对，也就是添加一个新元素。


```python
aDict = {'age': 35, 'name': 'Dong', 'sex': 'male'}
aDict['age'] = 39                  #修改元素值
aDict
```

```python
aDict['address'] = 'SDIBT'         #添加新元素
aDict
```

> ## 使用字典对象的update()方法可以将另一个字典的“键:值”一次性全部添加到当前字典对象，如果两个字典中存在相同的“键”，则以另一个字典中的“值”为准对当前字典进行更新。


```python
aDict = {'age': 37, 'score': [98, 97], 'name': 'Dong', 'sex': 'male'}
aDict.update({'a':97, 'age':39})  #修改’age’键的值，同时添加新元素’a’:97
aDict
```

## 3.4.2  集合操作与运算


> ## 使用集合对象的add()方法可以增加新元素，如果该元素已存在则忽略该操作，不会抛出异常；update()方法用于合并另外一个集合中的元素到当前集合中，并自动去除重复元素。

```python
s = {1, 2, 3}
s.add(4) 
s
```

```python
s.update({3,4}) 
s
```

> ## remove()方法用于删除集合中的元素，如果指定元素不存在则抛出异常；

> ## discard()方法用于从集合中删除一个特定元素，如果元素不在集合中则忽略该操作；

> ## clear()方法清空集合删除所有元素。


```python
s.pop()
```

```python
s.discard(3) #删除元素，不存在则忽略该操作
```

## 3.4.3  集合应用案例


## 问题解决：返回指定范围内一定数量的不重复数字。


```python
import random

def randomNumbers(number, start, end):
    '''使用集合来生成number个介于start和end之间的不重复随机数'''
    data = set() # 填空
    while len(data)<number:
        element = random.randint(start, end)
        data.add(element)#
    return data
```

## 问题解决：下面两段代码用来测试指定列表中是否包含非法数据，很明显第二段使用集合的代码更高效一些。


```python
import random

lstColor = ('red', 'green', 'blue')
lstColor1 = ('black', 'green', 'blue','yellow')
colors = [random.choice(lstColor1) for i in range(10000)]

for item in colors:                    #遍历列表中的元素并逐个判断
    if item not in lstColor:
        print('error:', item)
        break
#colors
```

```python
if (set(colors)-set(lstColor)):        #转换为集合之后再比较
    print('error')
```

```python
from random import randrange

# 其他用户喜欢看的电影清单
data = {'user'+str(i):{'film'+str(randrange(1, 10))\
                        for j in range(randrange(15))}\
        for i in range(10)}

# 待测用户曾经看过并感觉不错的电影
user = {'film1', 'film2', 'film3'}
# 查找与待测用户最相似的用户和Ta喜欢看的电影
similarUser, films = max(data.items(), key=lambda item: len(item[1]&user))
print('历史数据：')
for u, f in data.items():
    print(u, f, sep=':')
print('和您最相似的用户是：', similarUser)
print('Ta最喜欢看的电影是：', films)
print('Ta看过的电影中您还没看过的有：', films-user)
```

```python
from random import randrange
data = {'user'+str(i):({'film'+str(randrange(1, 10))\
                        for j in range(randrange(15))},{'music'+str(randrange(1, 10))\
                        for j in range(randrange(15))})\
        for i in range(10)}
# 待测用户曾经看过并感觉不错的电影以及听过感觉不错的音乐
user = ({'film1', 'film2', 'film3'},{'music1', 'music2', 'music3'})
similarUser, films_music = max(data.items(), key=lambda item: len(item[1][0]&user[0])+len(item[1][1]&user[1]))
print('和您最相似的用户是：', similarUser)
print('Ta最喜欢看的电影是：', films_music[0])
print('Ta最喜欢听的音乐是：', films_music[1])
print('Ta看过的电影中您还没看过的有：', films_music[0]-user[0])
print('Ta听过的音乐中您还没听过的有：', films_music[1]-user[1])
```
