# 7.3.1  使用%运算符进行格式化



# 格式符辅助符：

## '% [-] [+] [0] [m] [.n] 格式字符 '% x

## % ： 格式标志，表示格式开始

## [-] ：指定左对齐输出

## [+]：对正数加正号

## [0]：指定空位填0

## [m]：指定最小宽度

## [.n]：指定精度

## 格式字符：指定类型

## %：格式运算符

## x：待转换的表达式

# 常用格式字符：

```python
x = 1235
so = '%o' % x
so
```

```python
sh = "%x" % x
sh
```

```python
se = "%e" % x
se
```

# 7.3.2  使用format()方法进行格式化


```python
1/3
```

```python
print('{0:.3f}'.format(1/3,0.5)) #保留3位小数
```

```python
print('{1}'.format(1/3,0.5)) #保留3位小数
```

```python
'{0:%}'.format(3.5) #格式化为百分数
```

```python
'{0:_},{0:_x}'.format(1000000) #Python 3.6.0及更高版本支持
```

```python
'{0:_},{0:#x}'.format(1000000) #Python 3.6.0及更高版本支持
```

```python
print("The number {0:,} in hex is: {0:#x}, the number {1} in oct is {1:#o}".format(5555,55))
```

```python
 print("The number {1:,} in hex is: {1:#x}, the number {0} in oct is {0:o}".format(5555,55))
```

```python
print("my name is {name}, my age is {age}, and my QQ is {qq}".format(name = "Dong Fuguo",age = 40,qq = "30646****"))
```

```python
position = (5, 8, 13)
print("X:{0[0]};Y:{0[1]};Z:{0[2]}".format(position))
```

```python
position = [(5, 8, 13),('a','b','c')]
print("X:{0[0][0]};Y:{0[1][2]};Z:{0[0][2]}".format(position))
```

```python
position1 = (5, 8, 13)
position2 = ('a','b','c')
print("X:{0[0]};Y:{0[1]};Z:{1[1]}".format(position1,position2))
```

```python
weather = [("Monday","rainy"),("Tuesday","sunny"),
           ("Wednesday", "sunny"),("Thursday","rainy"),
           ("Friday","cloudy")]
formatter = "Weather of '{0[0]}' is '{0[1]}'".format
for item in map(formatter,weather):
    print(item)
```

```python
for item in weather: # 和上面等价
    print(formatter(item))
```

```python
type(formatter)
```

```python
f = lambda x:x+2 #类比
f(3)
```

# 7.4.1  find()、rfind()、index()、rindex()、count()


```python
s="apple,peach,banana,peach,pear"
s.find("peach")
```

```python
s.find("peach",7)
```

```python
s.find("peach",7,20)#  [start, end) #不存在抛出-1
```

```python
s.rfind('p')
```

```python
s.index('p')
```

```python
s.index('ppp')# 不存在抛出异常
```

```python
s="apple,peach,banana,peach,pear"
s.count('p'),s.count('pp'),s.count('ppp')
```

# 7.4.2  split()、rsplit()、partition()、rpartition()


```python
s = "apple,peach,banana,pear,"
s.split(",")
```

```python
s.partition(',')
```

```python
s = "apple,peach,banana,pear"
s.rpartition('banana')
```

```python
s = "2022-10-31"
t = s.split("-")
print(t)
```

```python
print(list(map(int, t)))
```

```python
s = 'hello world \n\n My name is Dong   '
s.split()
```

```python
s = '\n\nhello world \n\n\n My name is Dong   '
s.split()
```

```python
s = '\n\nhello\t\t world \n\n\n My name\t is Dong   '
s.split()
```

```python
s = '\n\nhello\t\t world \n\n\n My name is Dong   '
s.split(None, 1)          # 第一个参数None表示使用任意空白字符做分隔符 （跳过开头空白，找到第一个非空白后的分隔符分割，保留剩余内容（含空白））
```

```python
s = '\n\nhello\t\t world \n\n\n My name is Dong   '
s.rsplit(None, 1) 
```
