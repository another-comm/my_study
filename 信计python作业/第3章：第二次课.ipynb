# 3.1.7  切片操作的强大功能


### 功能：切片用来访问有序序列中的部分元素。

# 切片的功能展示
 

### 返回包含原列表中所有元素的新列表

```python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
bList = aList[::] 
bList, id(aList), id(bList)
```

```python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
bList = aList[9:3:-2] 
bList 
```

### 切片操作返回的列表是原来列表部分元素的浅复制

> 重点关注列表中的可变数据类型：列表，字典、集合（会互相影响）

```python
aList = [3, 4, [5,6]]
bList = aList[::]  
print(id(aList), id(bList),id(aList[2]), id(bList[2]))
bList.append(8)
aList, bList
```

```python
bList[2].append(10)
aList, bList
```

```python
aList = [3, 4, {5,6}]
bList = aList[::] 
bList[2].add(10)
aList, bList
```

```python
aList = [3, 4, {5:2,6:7}]
bList = aList[::] 
bList[2][7] = 8
aList, bList
```

```python
bList.append('hello')
bList, aList
```

### 返回包含原列表中所有元素的逆序列表

```python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
aList[::-1] 
```

### 隔一个取一个，获取偶(奇）数位置的元素

```python
aList[::2], aList[1::2] 
```

### 指定切片的开始和结束位置

```python
aList[3:6]
```

```python
#切片结束位置大于列表长度时，从列表尾部截断
aList[0:100] 
```

```python
#切片开始位置大于列表长度时，返回空列表
aList[100:]
```

```python
aList = [3, 5, 7]
aList[len(aList):]
```

```python
aList = [3, 4, 5, 6, 7, 9, 11, 13, 15, 17]
aList[-1:-2:-1] 
```

```python
aList=[3,4,5,6,7,9,11,13,15,17]
aList[-10:-9],aList[-1:-10:-1]
```

### （2）使用切片为列表增加元素

> ### 可以使用切片操作在列表任意位置插入新元素，不影响列表对象的内存地址，属于原地操作。


```python
aList = [1,1,1,1]
aList[len(aList):] = [9] # 在列表尾部增加元素
aList[:0] = [1, 2] #在列表头部插入多个元素
aList[3:3] = [4] #在列表中间位置插入元素
#aList.insert(3,4)
aList
```

### （3）使用切片替换和修改列表中的元素

```python
aList = [3, 5, 7, 9]
aList[:3] = [1, 2, 3, 4]#替换列表元素, 切片连续，等号两边的列表长度可以不相等
aList 
```

```python
aList[::2] = [0]*3 # 隔一个修改一个，切片不连续，两边元素要一样
aList 
```

```python
aList[::2] = ['a', 'b', 'c']
aList 
```

```python
aList=[1,2,3,4,5,6,7]
aList[1::2] = range(3)
aList
```

```python
aList=[1,2,3,4,5,6,7]
aList[1::2] = map(lambda x: x!=5, range(3))
aList
```

```python
aList=[1,2,3,4,5,6,7]
aList[1::2] = zip('abc', range(3))
aList
```

### (4) 使用切片删除列表中的元素

```python
aList = [3, 5, 7, 9]
aList[:3] = []                  # 删除列表中前3个元素，切片步长必须为1
aList
```

### 也可以结合使用del命令与切片结合来删除列表中的部分元素，此时切片可以不连续，不要求step必须为1

```python
aList = [3, 5, 7, 9, 11]
del aList[:3]                   # 切片元素连续
aList
```

```python
aList = [3, 5, 7, 9, 11]
del aList                   
aList
```
